Determine the KAT-7 system temperature using 

$T_{sys} = T_{bg}+T_{atm}(\theta) + T_{ant}(\theta)+T_{rx}+T_{cold}+T_{source}$

In this observation we have the Moon as one source and cold sky as the other. To reference the temp of the Moon to its antenna temperature we made use of equation (11) in Baars [1973] to get

$T_{source}=?$

For KAT-7 $D=12$m, $\eta_{A}=0.7$. So for the "hot load" we can use the temperature of the Moon as $T_b$. 

$T_{sys}'=T_{bg}+T_{atm}(\theta) + T_{ant}(\theta)+T_{rx}+T_{cold}$

so for $Y=P_{hot}/P_{cold}$ with the observation parameters the same for hot and cold we get

$Y = \frac{T_{bg}+T_{atm}(\theta) + T_{ant}(\theta)+T_{rx}+T_{cold}+T_{source}}{T_{bg}+T_{atm}(\theta) + T_{ant}(\theta)+T_{rx}+T_{cold}}$

$=\frac{T_{sys}'+T_{source}}{T_{sys}'}$ and

$T_{sys}' = \frac{T_{source}}{Y-1}$
