## P Pdot diagram

Rotational power loss:
$${dE_{\rm rot} \over d t} = {- 4 \pi^2 I \dot{P} \over P^3}\rlap{\quad}$$

Minimum magnetic field strength:
$$\biggl( {B \over {\rm Gauss}} \biggr) > 3.2 \times 10^{19} \biggl( { P \dot{P} \over {\rm s} } \biggr)^{1/2}\rlap{\quad
}$$

The characteristic age of a pulsar is defined by
$$ \tau \equiv { P \over 2 \dot{P}}\rlap{\quad}$$

useful for
following the lives of
pulsars, playing a role similar to the Hertzsprung-Russell diagram for
ordinary stars.&nbsp; It encodes a tremendous amount of information
about the pulsar population and its properties, as determined and
estimated from two of the primary observables, $P$ and $\dot P$.&nbsp;
Using those parameters, one can estimate the pulsar age, magnetic field
strength $B$, and spin-down power $\dot E$.&nbsp; (From the Handbook of
Pulsar Astronomy, by
Lorimer and Kramer)</span><br style="font-style: italic;">
</p>
<hr style="width: 100%; height: 2px; font-style: italic;">
<h3 style="font-family: verdana;">The Lives of Pulsars</h3>
<p style="font-family: verdana;">Pulsars are born in supernovae and
appear in the upper left corner
of the $P
\dot{P}$ diagram. If $B$ is conserved and they age as
described above, they gradually move to the right and down, along lines
of constant $B$ and crossing lines of constant characteristic age.
Pulsars with characteristic ages $ &lt; 10^5$ yr are often found in
or near recognizable supernova remnants. Older pulsars are not, either
because their SNRs have faded to invisibility or because the supernova
explosions expelled the pulsars with enough speed that they have since
escaped from their parent SNRs. The bulk of the pulsar population is
older than $10^5$ yr but much younger than the Galaxy ($\sim
10^{10}$ yr). The observed distribution of pulsars
in the $P \dot{P}$ diagram indicates that something changes as pulsars
age.&nbsp; One controversial possibility is that the
magnetic fields of old pulsars must
decay on time scales $\sim 10^7$ yr, causing old pulsars to move almost
straight down in the $P \dot{P}$ diagram until they fall
below into the graveyard below the <span
 style="font-style: italic; font-weight: bold; color: rgb(204, 0, 0);">
death line</span><span style="font-style: italic;"></span> and cease
radiating radio pulses.<br>
</p>
<p style="font-family: verdana;">Almost all short-period pulsars below
the
<span
 style="font-style: italic; font-weight: bold; color: rgb(204, 0, 0);">spin-up
line</span> near $\log
[\dot{P}/P({\rm sec})] \approx -16$ are in binary systems, as evidenced
by periodic (i.e.
orbital) variations
in their observed pulse periods. These <span
 style="font-style: italic; font-weight: bold; color: rgb(204, 0, 0);">
recycled pulsars</span> have
been spun up by accreting mass
and angular momentum from their companions, to the point that they emit
radio pulses despite their relatively low magnetic field strengths $B
\sim 10^8$ G (the accretion causes a substantial reduction in the
magnetic field strength). The magnetic fields of neutron stars funnel
ionized
accreting material onto the magnetic polar caps, which become so hot
that they emit X-rays. As the neutron stars rotate, the polar caps
appear and disappear from view, causing periodic fluctuations in X-ray
flux; many are detectable as X-ray pulsars.
</p>
<p style="font-family: verdana;"><span
 style="font-style: italic; font-weight: bold; color: rgb(204, 0, 0);">Millisecond
pulsars</span> (MSPs) with
low-mass ($M \sim 0.1-1
M_\odot$) white-dwarf companions
typically have orbits with small eccentricities.&nbsp; Pulsars with
extremely
eccentric orbits usually have neutron-star companions, indicating that
these
companions also exploded as supernovae and nearly disrupted the binary
system.&nbsp; Stellar interactions in globular clusters cause a much
higher fraction of recycled pulsars per unit mass than in the Galactic
disk.&nbsp; These interactions can result in very strange systems such
as pulsar&#8211;main-sequence-star binaries and MSPs in highly eccentric
orbits.&nbsp; In both cases, the original low-mass companion star that
recycled the pulsar was ejected in an interaction and replaced by
another star.&nbsp; (The <span
 style="font-style: italic; font-weight: bold; color: rgb(204, 0, 0);">eccentricity</span>
$e$ of an elliptical orbit is defined as the ratio of the separation of
the foci to the length of
the major axis.&nbsp; It ranges between $e =0$ for a circular orbit and
$e = 1$ for a parabolic orbit.)<br>
</p>
<p style="font-family: verdana;">A few millisecond pulsars are
isolated. They were probably recycled via the standard scenario in
binary systems, but the energetic millisecond pulsars eventually
ablated their companions away.

## Pulsar Sensitivity

The conversion from temperature to flux density involves the gain introduced by the collecting area of the dish itself, referred to as *antenna gain*, which is given by

\begin{equation}
  G = \frac{\eta A}{2 \text{k}} \times 10^{-26} \approx \frac{\eta A}{2761}, \quad \text{[K / Jy]}
\end{equation}

where $\eta$ is the efficiency of the dish (less than 1!), $A$ is the dish area in square metres and $\text{k}$ is Boltzmann's constant.

- Estimate the antenna gain for a KAT-7 dish with a diameter of 12 metres and an efficiency of 0.5 (50%)

#### Example

```python
# Area of KAT-7 dish

D = 12                                      # diameter of dish
e = 0.5                                     # efficiency 

A = numpy.pi * (D/2)**2                     # calculate area of KAT7 dish
    
G = (e * A) / 2761

print "Gain = ",G
```

The minimum detectable flux, $S_{min}$ in mJy for a pulsar with period $P$ and pulse width $W$ is

\begin{equation}
S_{min} = \dfrac{\frac{S}{N}_{min} T_{sys}}{G  N_{ant}  \sqrt{n_{p}t_{int} B} }\sqrt{\dfrac{W}{P-W}}
\end{equation}

where

* $N_{ant}$ is the number of antennas
* $n_{p}$ is the number of polarizations
* $B$ is the observing bandwidth
* $t_{int}$ is the length of the observation
* $S/N_{min}$ is the detection threshold signal-to-noise

Assume a detection threshold of $S/N_{min} = 10$



### KAT-7 parameters

You can find the parameters for KAT-7 on 

http://public.ska.ac.za/kat-7

```python
#Antenna array properties
Nant = 7   # 7 antennas
D = 12     # diameter of dish
e = 0.5    # efficiency 

A = numpy.pi * (D/2)**2                     # calculate area of KAT7 dish

G = (e * A) / 2761                          # Gain

print "Gain = ",G
# Backend properties

Tsys = 30  # system temperature is 30 K
B = 200    # bandwidth 200 MHz
Np = 2     # 2 polarizations
```

Calculate the minimum detectable flux for a pulsar with period $P = 0.2 {\rm s}$, pulse width $W = 6 {\rm ms} $?

Assume the observation is 15 min long?

```python
P = 0.2  # pulsar period
W = 6.0E-3 # pulse width in seconds
Tint = 15*60.0 # integration length

Smin = ((10*Tsys)/(Nant*G*numpy.sqrt(Np*B*Tint)))*numpy.sqrt(W/(P-W))

print "Smin = ",Smin, " mJy"
```

## ATNF pulsar catalogue

The ATNF pulsar catalogue contains details of all published pulsars.

http://www.atnf.csiro.au/people/pulsar/psrcat/

I have downloaded all of the pulsars visible from KAT-7 - those that appear above an elevation of 15 degrees.



```python
import numpy as np
from matplotlib import pylab as plot
#%matplotlib inline

plscatfile='psrcat201410.csv'

#load the catalogue into the numpy array pls

pls=np.loadtxt(plscatfile,unpack=True,dtype=str)
```

The table consists of the following columns

    0. PSRJ: Pulsar Jname
    1. P0 Barycentric period of the pulsar (sec)
    2. P1	Period derivative
    3. W50	Width of pulse at 50% (msec)
    4. S1400 Flux at 1400MHz (mJy)
    5. DM Dispersion measure (cm^{-3} pc)
    6. RAJD Right Ascension (J2000) (deg)
    7. DECJD Declination (J2000) (deg)
    8. GL	Galactic Longitude (deg) 
    9. GB	Galactic Latitude (deg)
    10. BINARY Binary model



```python
# read the columns into variables

NAME =pls[0]
P = np.array(pls[1],dtype = float)
Pdot = np.array(pls[2],dtype=float)
W50s = np.array(pls[3],dtype=float)*1E-3
Flux = np.array(pls[4],dtype=float)
DM = np.array(pls[5],dtype=float)
RA = np.array(pls[6],dtype=float)
DEC = np.array(pls[7],dtype=float)
GLONG = np.array(pls[8],dtype=float)
GLAT = np.array(pls[9],dtype=float)
BIN = np.array(pls[10]!='*',dtype=int)

NPossible = len(NAME)


```

```python
PulsarName = 'J0835-4510'


print "Flux = ", Flux[NAME==PulsarName][0], " mJy"
print "Period = ", P[NAME==PulsarName][0]," s"
print "Pulse width = ",W50s[NAME==PulsarName][0]*1000, " ms"
```

## Pulsars observable with KAT-7

Plot out the pulsars in the P-Pdot diagram that can be observed with KAT-7 in 1 hours of observing time.


You can find the parameters for KAt-7 on 

http://public.ska.ac.za/kat-7

```python
# Observing time

Tint = 15 * 60

#Antenna array properties
Nant = 7   # 7 antennas
D = 12     # diameter of dish
e = 0.5    # efficiency 

A = np.pi * (D/2)**2                     # calculate area of KAT7 dish

G = (e * A) / 2761                          # Gain

# Backend properties

Tsys = 30  # system temperature is 30 K
B = 200    # bandwidth 200 MHz
Np = 2     # 2 polarizations


# calculate minimum detectable flux in mJy for each source in the catalogue
# use the values for the period and pulse width from the catalogue


Smin = ((10*Tsys)/(Nant*G*np.sqrt(2*B*Tint)))*np.sqrt(W50s/(P-W50s))

observable = np.where(Flux>=Smin)
notobservable = np.where(Flux<Smin)




```

```python
#Set up a log-log plot for P vs Pdot
figure=plt.figure(figsize=(15,10))
plt.title("Observable")
plt.ylabel("log Pdot")
plt.xlabel("log (Period/[s])")
plt.yscale('log')
plt.xscale('log')
plt.plot(P[observable],Pdot[observable],'g.')
plt.plot(P[notobservable],Pdot[notobservable],'r.')
```

## Sensivity curves

To get a better idea of the affect of observing parameters on the number of observable pulsars we can plot some different 
sensivity curves.

### Plot fraction of pulsars observable as a function of observation length

Plot the observable fraction as a function of observing time.

```python
# calculate fraction of pulsars observable as a function of observing time

Tvals = np.arange(10,3610,10)    # array containing integration times

ObsFraction=[]

for index in range(len(Tvals)):
        T_i = Tvals[index]
        
        # calculate the minimum detectable flux for each pulsars assuming an integration time of T_i
        Smin_i = ((10*Tsys)/(Nant*G*np.sqrt(2*B*T_i)))*np.sqrt(W50s/(P-W50s))
        
        # array with all pulsars with flux greater than minimum detectable flux
        
        obs_i = np.where(Flux>=Smin_i)
   
        # calculate fraction of pulsars that are observable to array
        ObsFraction.append((len(obs_i[0])/float(NPossible)))


     
```

```python
#Set up a log-log plot for P vs Pdot
figure=plt.figure(figsize=(15,10))
plt.title("Fraction of pulsars observable as a function of observing time")
plt.ylabel("Observable fraction")
plt.xlabel("Integration time /[s])")
plt.plot(Tvals,ObsFraction,'g.')


```

### Calculate minimum detectable flux for a 100 ms pulsar

```python
# calculate fraction of pulsars observable as a function of observing time

Tvals = np.arange(10,3610,10)    # array containing integration times


# calculate the minimum detectable flux for each a pulsar with P = 0.1 s and W = 0.03 s
# assuming an integration time of Tval
          
Flux100ms = ((10*Tsys)/(Nant*G*np.sqrt(Np*B*Tvals)))*np.sqrt(0.03/(0.1-0.03))
```

```python

figure=plt.figure(figsize=(15,10))
plt.title("Observable")
plt.ylabel("Minimum detectable flux for 100 ms pulsar")
plt.xlabel("Integration time /[s])")
plt.plot(Tvals,Flux100ms,'g.')
```
